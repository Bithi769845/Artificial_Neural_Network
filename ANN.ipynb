```python
df=pd.read_csv('/content/gdrive/MyDrive/Dataset/ddos_dataset_sdn.csv')
df
```

```python
#Handle null values
Mean = df.rx_kbps.mean()
Mean
df.rx_kbps = df.rx_kbps.fillna(Mean)
```

```python
Mean = df.tot_kbps.mean()
Mean
df.tot_kbps = df.tot_kbps.fillna(Mean)
```

```python
Nul = df.isnull().sum()
```

```python
#One Hot Encoding
dummy_variables = pd.get_dummies(df['Protocol'],drop_first=True)
dummy_variables.head()
new_df=df.drop('Protocol',axis=1)
new_df.head()
df=pd.concat([new_df,dummy_variables],axis=1)
df.head()
```

```python
dummy_variables = pd.get_dummies(df['src'],drop_first=True)
dummy_variables.head()
new_df=df.drop('src',axis=1)
new_df.head()
df=pd.concat([new_df,dummy_variables],axis=1)
df.head()
```

```python
dummy_variables = pd.get_dummies(df['dst'],drop_first=True)
dummy_variables.head()
new_df=df.drop('dst',axis=1)
new_df.head()
df=pd.concat([new_df,dummy_variables],axis=1)
df.head()
```

```python
x=df.drop(['label'],axis=1)
x
```

```python
y=df['label']
y
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=.3,random_state=1)
```

```python
#Performing Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
#Initialising ANN
ann = tf.keras.models.Sequential()
```

```python
#Adding First Hidden Layer
ann.add(tf.keras.layers.Dense(units=40,activation="relu"))
```

```python
#Adding Second Hidden Layer
ann.add(tf.keras.layers.Dense(units=20,activation="relu"))
```

```python
#Adding Output Layer
ann.add(tf.keras.layers.Dense(units=1,activation="sigmoid"))
```

```python
#Compiling ANN
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
```

```python
#Fitting ANN
ann.fit(xtrain,ytrain,batch_size=32,epochs = 100)
```

```python
y_pred = ann.predict(xtest)
y_pred = (y_pred > 0.5)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(ytest, y_pred)
print(cm)
accuracy_score(ytest,y_pred)
```
